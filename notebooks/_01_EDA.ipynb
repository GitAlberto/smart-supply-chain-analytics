```python
# Sélection des colonnes retenues
df_c = df[[
    # Délais & Livraison
    'Days for shipping (real)',
    'Days for shipment (scheduled)',
    'Delivery Status',
    'Late_delivery_risk',
    'Shipping Mode',
    'order date (DateOrders)',
    'shipping date (DateOrders)',
    # Géographie
    'Market',
    'Order Region',
    'Order Country',
    # Client
    'Customer Segment',
    # Commande & Produit
    'Order Status',
    'Category Name',
    'Product Name',
    'Order Item Quantity',
    # Finance
    'Sales',
    'Benefit per order',
    'Order Profit Per Order',
    'Order Item Profit Ratio'
]]

print(df_c.shape)
```

```python
df_c.head(10)
```

```python
# Renomage des colonnes
df_c = df_c.rename(columns={
    'order date (DateOrders)'        : 'order_date',
    'shipping date (DateOrders)'     : 'dispatch_date',
    'Days for shipping (real)'       : 'delivery_days_real',
    'Days for shipment (scheduled)'  : 'delivery_days_scheduled',
    'Delivery Status'                : 'delivery_status',
    'Late_delivery_risk'             : 'late_delivery_risk',
    'Shipping Mode'                  : 'shipping_mode',
    'Market'                         : 'market',
    'Order Region'                   : 'order_region',
    'Order Country'                  : 'order_country',
    'Customer Segment'               : 'customer_segment',
    'Order Status'                   : 'order_status',
    'Category Name'                  : 'category_name',
    'Product Name'                   : 'product_name',
    'Order Item Quantity'             : 'quantity',
    'Sales'                          : 'sales',
    'Benefit per order'              : 'benefit_per_order',
    'Order Profit Per Order'         : 'profit_per_order',
    'Order Item Profit Ratio'        : 'profit_ratio'
})

print(df_c.columns.tolist())
```

```python
### Pourcentage des valeurs manquantes du dataset ###
nb_cells = df_c.shape[0]*df_c.shape[1]
print("Nombre de cellules : ", nb_cells)

Pct_nan = (df_c.isna().sum().sum() / nb_cells)*100
print("Pourcetage de valeurs manquantes : ", Pct_nan)
```

<h4>Traitement des dates<h4>

```python
# Conversion des dates en datetime
df_c['order_date']    = pd.to_datetime(df_c['order_date'])
df_c['dispatch_date'] = pd.to_datetime(df_c['dispatch_date'])
```

```python
# Calcul des dates de livraison
df_c['delivery_date_real']      = df_c['dispatch_date'] + pd.to_timedelta(df_c['delivery_days_real'], unit='d')
df_c['delivery_date_scheduled'] = df_c['dispatch_date'] + pd.to_timedelta(df_c['delivery_days_scheduled'], unit='d')
```

```python
# Vérification
df_c[['dispatch_date', 'delivery_days_real', 'delivery_date_real', 'delivery_date_scheduled']].head(5)
```

```python
# Suppression des doublons
df_c = df_c.drop_duplicates()

# Vérification
print("Doublons :", df_c.duplicated().sum())
print("Shape après suppression :", df_c.shape)
```
